# Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2: Load Dataset

```python
# Load PIMA Diabetes Dataset from GitHub (UCI version)
url = 'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv'
df = pd.read_csv(url)

df.head()
```

# Step 3: Data Overview

```python
df.info()
df.describe()
```

# Step 4: Glucose Level vs Age

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=df)
plt.title('Glucose Level vs Age')
plt.show()
```

# Step 5: BMI Patterns

```python
plt.figure(figsize=(8,5))
sns.histplot(df['BMI'], bins=30, kde=True)
plt.title('BMI Distribution')
plt.show()
```

# Step 6: Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Step 7: BMI Distribution by Outcome (Boxplot)


```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI by Outcome')
plt.show()
```

# Step 8: Age Distribution by Outcome (Violin Plot)


```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Outcome', y='Age', data=df, palette='muted')
plt.title('Age Distribution by Outcome')
plt.show()
```

# Step 9: Diabetes Outcome Counts (Countplot)


```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df, palette='pastel')
plt.title('Diabetes Outcome Counts')
plt.show()
```

# Step 10: Average Glucose by Age (Lineplot)


```python
age_glucose = df.groupby('Age')['Glucose'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x='Age', y='Glucose', data=age_glucose, color='red')
plt.title('Average Glucose by Age')
plt.show()
```

# Conclusion
The Diabetes Risk Analysis project highlights patterns in the **PIMA Diabetes Dataset**.  

- Higher glucose levels show a correlation with positive diabetes outcomes.  
- BMI distribution indicates higher risk in overweight individuals.  
- Heatmaps reveal strong correlations between glucose, BMI, and diabetes outcome.
